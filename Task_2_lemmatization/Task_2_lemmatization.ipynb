```python
import nltk
nltk.download('stopwords')
```

```python
# Функция, определяющая, является ли строка пустой
def isBlank (string):
    if string and string.strip():
        return False
    return True
```

```python
import re

patterns = "[A-Za-z0-9(){}–«»,:;\"'.“”!#$%&<=>?@\xa0/№-]+"

texts = ''

# В цикле читаются все документы из папки ./documents, из текста каждого документа убираются спец. символы,
# указанные в переменной patterns
# После этого текст всех документов собирается в одну строковую переменную
for i in range(1,101):
    path='./documents/out{num}.txt'.format(num=i)
    with open(path,'r') as file:
        text = ''
        for line in file.readlines():
            if not isBlank(line):
                line = re.sub(patterns, ' ', line)
                text += ' ' + line
        texts += ' ' + text

print(texts)
```

```python
from pymorphy2 import MorphAnalyzer
from nltk.corpus import stopwords

stopwords_ru = stopwords.words("russian")
morph = MorphAnalyzer()

# Функция для получения словаря лемм, где ключ - лемма, значение - массив слов, относящихся к этой лемме
# Строка разбивается на слова по пробелам, проверяется, является ли слово стоп-словом,
# каждое слово добавляется в сет слов. 
# После этого каждое слово приводится к нормальной форме и добавляется в словарь как ключ, 
# а само слово добавляется в массив слов по этому ключу

def lemmatize(text):
    lemmas = {}
    tokens = set()
    for word in text.split(' '):
        token = word.lower()
        if token and token not in stopwords_ru:
            token = token.strip()
            if len(token) > 1:
                tokens.add(token)
            nor_token = morph.normal_forms(token)[0]
            if len(nor_token) > 1:
                if nor_token in lemmas.keys() and token not in lemmas[nor_token]:
                    lemmas[nor_token].append(token)
                else:
                    lemmas[nor_token] = [token]
    if len(tokens) > 2:
        return tokens,lemmas
    return None


tks,lemmas = lemmatize(texts)
print(lemmas.keys())
```

```python
# Все слова записываются в файл tokens.txt
with open('./result/tokens.txt', 'w') as f:
    for t in tks:
        f.write('{}\n'.format(t))

                
```

```python
# Все леммы с относящимися к ним словами записываются в файл lemmas.txt
with open('./result/lemmas.txt', 'w') as f:
    for key,value in lemmas.items():
        ts = key + ': ' + ' '.join(value)
        f.write('{}\n'.format(ts))
        
```
